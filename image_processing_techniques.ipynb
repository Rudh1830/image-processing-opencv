```python

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# -----------------------------
# READ IMAGE AND DISPLAY DETAILS
# -----------------------------

image_path = '/content/teddy.png'
image = cv2.imread(image_path)

# Safety check
if image is None:
    raise ValueError("Image not loaded. Check the image path.")

print("Image type:", type(image))        # numpy.ndarray
print("Image shape:", image.shape)       # (height, width, channels)

# Display original image
cv2_imshow(image)

# -----------------------------
# ACCESSING COLOR CHANNELS
# -----------------------------

# OpenCV uses BGR format
red_channel = image[:, :, 2]

print("Red channel shape:", red_channel.shape)  # 2D array
cv2_imshow(red_channel)

# -----------------------------
# GRAYSCALE CONVERSION
# -----------------------------

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

print("Gray image shape:", gray_image.shape)
cv2_imshow(gray_image)

# -----------------------------
# REMOVE RED CHANNEL (CORRECTED)
# -----------------------------

image_no_red = image.copy()
image_no_red[:, :, 2] = 0   # Removing RED channel
cv2_imshow(image_no_red)

# -----------------------------
# IMAGE CROPPING
# -----------------------------

x, y = 350, 350
width, height = 500, 500

# Ensure cropping stays inside image boundaries
cropped_image = image[y:y+height, x:x+width]
cv2_imshow(cropped_image)

# -----------------------------
# GAUSSIAN BLUR
# -----------------------------

blurred_image = cv2.GaussianBlur(image, (125, 125), 0)
cv2_imshow(blurred_image)

# -----------------------------
# IMAGE SHARPENING
# -----------------------------

kernel = np.array([
    [-1, -1, -1],
    [-1,  9, -1],
    [-1, -1, -1]
])

sharpened_image = cv2.filter2D(image, -1, kernel)
cv2_imshow(sharpened_image)

# -----------------------------
# ADAPTIVE THRESHOLDING
# -----------------------------

adaptive_threshold = cv2.adaptiveThreshold(
    gray_image,
    255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    115,
    1
)

cv2_imshow(adaptive_threshold)
```
